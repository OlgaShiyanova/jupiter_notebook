##### 1.3. Изучите распределение квадратуры жилой

```python
plt.figure(figsize=(5,4))
plt.hist(df['long'], color='#AA0011', edgecolor='black')
plt.title('распределение квадратуры')
plt.xlabel('m 2')
plt.ylabel('количество')
```

```python
plt.figure(figsize=(15,10))
plt.scatter(df['long'], df['lat'], c=df['price'], cmap='viridis')
plt.colorbar(label='Цена')
plt.title('Распределение квадратуры')
plt.xlabel('Длина')
plt.ylabel('Ширина')
plt.grid(True)
plt.show()
```

```python
data = df['floors'].value_counts()
data
```
