## решение одногрупника

````python
import pandas as pd

df = pd.read_csv('./Libs/kc_house_data.csv')

# среднюю стоимость домов в зависимости от количества спален и сохраните в avg.
# Отсортируйте от меньшей стоимости к большей.

av = df.groupby('bedrooms').agg({'price':'mean'})
avg = av.sort_values(by='price', ascending=True)
# print(avg)
# print(avg.info())

# Найдите минимальную min, среднюю mean и максимальную max стоимости домов в зависимости от состояния дома и
# сохраните в price.

price = df.groupby('condition').agg({'price': ['min', 'mean','max']})
# print(price)

# Постройте таблицу с подсчетом количества домов в данных в зависимости от вида на набережную waterfront и
# оценкой вида view и сохраните ее в view_waterfront.

view_waterfront = df.pivot_table(index='waterfront', columns='view', values='zipcode', aggfunc='count', fill_value=0)
# print(view_waterfront)

# Каких домов в зависимости от этажности и количества спален больше? Постройте эту таблицу, содержащую в себе
# информацию о спальнях и этажности, и сохраните ее в bedrooms_floors.

bedrooms_floors = df.pivot_table(index='floors', columns='bedrooms', values='zipcode', aggfunc='count', fill_value=0)
# print(bedrooms_floors)

# Постройте таблицу с подсчетом медианной стоимости домов в данных в зависимости от состояния дома
# и оценки дома и сохраните в 'median_price'.

median_price1 = df.groupby(['grade', 'condition']).agg({'price': 'median'}).unstack('grade')
median_price1.columns = median_price1.columns.droplevel(0)
median_price1 = median_price1.fillna(0)
median_price = median_price1.astype(int)
# print(median_price)
````

##  идеальное решение от автотеста

````python
import pandas as pd
df = pd.read_csv('./Libs/kc_house_data.csv', sep=',')

# 1. Средняя стоимость домов в зависимости от количества спален (отсортирована от меньшей к большей)
avg = df.groupby('bedrooms').agg({'price':'mean'}).sort_values('price')

# 2. Минимальная, средняя и максимальная стоимости домов в зависимости от состояния дома
price = df.groupby('condition').agg({'price':['min', 'mean', 'max']})

# 3. Таблица с подсчетом количества домов в зависимости от видов на набережную и оценок вида
view_waterfront = df.pivot_table(index='waterfront', columns='view', values='condition', aggfunc='count', fill_value=0)

# 4. Таблица с подсчетом количества домов в зависимости от этажности и количества спален
bedrooms_floors = pd.crosstab(index=df['floors'], columns=df['bedrooms'])

# 5. Таблица с медианной стоимостью домов в зависимости от состояния дома и оценки дома
median_price = df.pivot_table(index='condition', columns='grade', values='price', aggfunc='median', fill_value=0)
````
